Arrays and Lists

Sets are highly optimised to provide fast access to specific elements. Lists are optimised for storing elements in order.

Addresses usually start with 0x as they are hexadecimal

ARRAYS
 - A series of elements a1, a2, a3 usualy denoted as A[1], A[2] etc
 - Stored consecutively in memory (Contiguous)
 - All the same type (Homogenous)
 - Finding the ith element, can be done easily if the address of the beginning of the array is known. 
 - This is done with 0x-- + (i-1)w where w is ____
 - As these are arithmetic operations, finding the ith element is O(1)
 - Erasing an element, requires the all subsequent elements to be moved into the immediate lower addresses
 - This is very inefficient, and is in O(n)
 - Array size is basically fixed, as there is a lot of hassle related to moving it when the size increases, so twice the memory is allocated for it when it is moved, so that hopefully if it increases in size again it does not have to be moved

 LINKED LISTS
  - Lists are made up of nodes, each of which stores an element and a pointer to another node
  - The first node is referred to as the head
  - The last node is referred to as the tail and points to null
  - The nodes are not necessarily stored consecutively in memory and may be scattered across it, but this is less of an issue because each node points to the next, so they can still be found
  - We assume that for a given list L we have pointers to the head (L.head) and tail (L.tail), and maybe to a variable holding the size (L.size) of the list
  - For a node N, N.data is the element and N.next is the pointer

INPUT: list L, positive int i
OUTPUT: ith piece of data in L
if i > L.size then
    return Out of Range
end if
N = L.head
for i = 0 to i do
    N = N.next
end for
return N.data

finding ith element of a list is O(i) time.

Note on Arrays
 - In the RAM model, Arrays may store integers of any size
 - |3|4|2|1|NULL||| (0x123 - 0x129)
 - To insert a number, we require address 0x128 to be open, and it will take moving several pieces of the array
 - If 0x128 is not available, the whole array may have to be moved to a place in memory with enough consecutive empty memory slots, this time reserved with twice the required memory

INSERTION AND DELETION IN LINKED LISTS
 - When inserting into a list, the element previous to the new one must have its .Next copied to the .Next of the new element, and then updated to point to the new element
 - To delete the head, we change the head pointer from N1 to N2. (L.Head = L.Head.Next). We may also need to do L.size = L.size - 1, if a size attribute exists. If the L.Head = L.Tail, we may also need to update the tail pointer as well (L.Head = Null, L.Tail = Null, L.size = 0). We should also ensure that the list is not empty, before attempting to remove from an emty list

Arrays: Fast data access, slow insertion/deletion
Linked Lists: slow data access, fast insertion/deletion
