Concurrent Version Systems (CVS)
 - Add/remove files, used on top of RCS

ABOUT GIT
 - Distributed Version Control System
 - Made by Linus Torvalds to manage Linux Kernel
 - Designed for speed
 - Wideley used, about 80.5%
 - Does not store diffs
 - No central repo, can synchronise with remote repos
 - Cloud hoted repo servers; github, gitlab and bitbucket

KEY CONCEPTS
 - A file (in a path)
 - A commit; snapshot of files in a state
 - A branch is a linear sequence of commits
 - A repository is a group of branches in a project
 - A remote is a different location where a repo is stored

GIT COMMANDS
Used on the command line, as 80% of devs do
 - "git init"
 Creates .git directory, where everything is stored. Possibly also at this stage use git config and maybe create .gitignore file
 - "git add"
 Current working file version into staging area, preparing for a commit
 "git add ." adds the whole directory
 - "git status"
 Checks what will be committed
 - "git commit"
 Makes commit based on currently staged files, starts an editor
 git commit -m "message" will bypass editor
 Leave editor with esc+:+w+q+!
 - "git push"
 Push a branch to a remote repo. "git push origin main" will push to default branch, origin is defined by the command "git remote add origin" or "git clone"
