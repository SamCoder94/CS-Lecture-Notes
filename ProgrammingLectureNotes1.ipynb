Version Control
Revision Control System (RCS)
 - Locked files so one person changed at a time
Concurrent Version Systems (CVS)
 - Add/remove files, used on top of RCS
Word Track Changes
Subversion
 - github before github
 - Bad to keep several file versions due to storage, so since difference between 2 files is less than whole file, use computation to store new file and difference from old
 - GITH and PATCH
git
 - Stores all of the files
 - This allows for merging branches, by designating one as the main branch, applying a backwards patch to the most recent version and then an alternate patch to the variant file
Mercurial
 - Growing to rival git
darcs

ABOUT GIT
 - Distributed Version Control System
 - Made by Linus Torvalds to manage Linux Kernel
 - Designed for speed
 - Wideley used, about 80.5%
 - Does not store diffs
 - No central repo, can synchronise with remote repos
 - Cloud hoted repo servers; github, gitlab and bitbucket

KEY CONCEPTS
 - A file (in a path)
 - A commit; snapshot of files in a state
 - A branch is a linear sequence of commits
 - A repository is a group of branches in a project
 - A remote is a different location where a repo is stored

GIT COMMANDS
Used on the command line, as 80% of devs do
 - "git init"
 Creates .git directory, where everything is stored. Possibly also at this stage use git config and maybe create .gitignore file
 - "git add"
 Current working file version into staging area, preparing for a commit
 "git add ." adds the whole directory
 - "git status"
 Checks what will be committed
 - "git commit"
 Makes commit based on currently staged files, starts an editor
 git commit -m "message" will bypass editor
 Leave editor with esc+:+w+q+!
 - "git push"
 Push a branch to a remote repo. "git push origin main" will push to default branch, origin is defined by the command "git remote add origin" or "git clone"

<h2>THE BOX MODEL</h2>
<p>CSS becomes easier when you view all elements of a page as some nested boxes</p>
<p>Box properties are not inherited from parent boxes. They can apply  to a whole box or individually to any of the sides</p>
<p>There is shorthand css to make this easier</p>
<p>The border property, has sub properties width, style and color</p>
<p>In shorthand, border properties apply in top right bottom left order</p>
<p>The padding property, is the space between the outside of the content and the inside of the border. It can be left empty or filled with background styles</p>
<p>Margin creates space between 2 boxes</p>
<p>Paddding and margin are used to manage spacing of content</p>
<p>Width can be used to limit how far across the website a piece of content can span</p>
<p>Background properties affect everything behind the content, inside the border</p>
<h2>Positioning</h2>
<p>Good CSS layout relies on good (X)HTML and is easier with a well structured content</p>
<p>Divs, Spans, Sections, classes and ids are essential for this</p>
<p>HTML vs XHTML doesn't matter as long as it is strictly one or the other</p>
<p>The natural page flow is the source order of the html element boxes</p>
<p>Final rendered position of each block determines the later ones</p>
<p>Browser parses html and css before drawing. Final appearance determined by CSS properties and browser defaults</p>
<p>CSS REpositions content</p>
<p>Float property allows later content to wrap around, but their blocks do not change shape</p>
<p>Clear property stops floating content from covering late content</p>
<p>None floated boxes may be turned into columns, which are fairly easy to create using floats</p>
<p>Static - default non-positioned value
Relative  - Element in page flow, but shown off set
Absolute - removed from page flow and placed anywhere
Fixed - removeed from page flow and locked to browser view port</p>
<p>Elements acting as origins for absolute positioning must themselves have a postion</p>
<p>Fixed positioning allows rest of content to scroll behind so is useful for nav bars</p>
